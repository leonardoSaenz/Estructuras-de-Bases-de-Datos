```python
import hashlib

# Función para calcular SHA-256
def hash_data(data):
    return hashlib.sha256(data.encode('utf-8')).hexdigest()

# Función que genera Merkle Root y compara con el root conocido
def verify_merkle_root(transactions, known_root):
    # Calcular hash de las transacciones (hojas)
    hashes = [hash_data(tx) for tx in transactions]
    print("Nivel 0 (hojas):", hashes)

    level = 0
    # Construir árbol respetando pares e impares
    while len(hashes) > 1:
        new_level = []
        i = 0
        while i < len(hashes):
            if i + 1 < len(hashes):
                # Combinar dos hashes
                combined = hashes[i] + hashes[i+1]
                new_hash = hash_data(combined)
                new_level.append(new_hash)
                print(f"Nivel {level+1}: hash({hashes[i]} + {hashes[i+1]}) = {new_hash}")
                i += 2
            else:
                # Nodo impar sube directo
                new_level.append(hashes[i])
                print(f"Nivel {level+1}: nodo impar sube = {hashes[i]}")
                i += 1
        hashes = new_level
        level += 1

    # hashes[0] es el Merkle Root calculado
    print("\nMerkle Root calculado:", hashes[0])
    if hashes[0] == known_root:
        print(" Coincide con el Merkle Root conocido")
        return True
    else:
        print(" No coincide con el Merkle Root conocido")
        return False

# ----------------------
# Ejemplo de uso
transactions = [
    "a",
    "b",
    "c",
    "d"
]

# Root hash que ya conoces
known_root = "58c89d709329eb37285837b042ab6ff72c7c8f74de0446b091b6a0131c102cfd"  # Pones tu hash real aquí

# Verificar
verify_merkle_root(transactions, known_root)
```
